=> From the above graph, you can see that Jan 2021 sales data was fluctuating between 0.5 to 2M but after that sales increased drasctically and then as time goes sales started fluactuating again but more than before jan 2021

##)) Detect and list outliers in sales using IQR method

```python
Q1 = data["Sales"].quantile(0.25)
Q3 = data['Sales'].quantile(0.75)
IQR = Q3-Q1
outliers = data[(data['Sales']<(Q1-1.5*IQR)) | (data['Sales']>(Q3+1.5*IQR))]
print(f"Number od outliers in the sales: {len(outliers)}")
display(outliers[['Date','Country','Brand','Sales']].sort_values(by='Sales',ascending=False).head(10))
```

=> we can see that most of the outliers are from Apple brand, suggesting that Apple might dominate the top end of the sales distribution.Countries like Brasil, Colombia, USA appers many time indicating that high sales are concentrating in few regions
